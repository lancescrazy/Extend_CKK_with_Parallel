```python
import re
import numpy as np
import pandas as pd
```

```python
files = ['norm.txt', 'para.txt', 'para_cut_20.txt', 'para_cut_30.txt', 'para_cut_40.txt']
```

```python
data = []
for file in files:
    with open(file) as f:
        readlines = f.readlines()
    read = []
    tmp = []
    for line in readlines:
        
        if 'numbers of integer' in line:
            tmp.append(file[:-4])
            num = re.findall("\d+",line)[0]
            tmp.append(num)
        if 'totalmax of 10 trials' in line:
            num = re.findall("\d+",line)[1]
            tmp.append(str(int(num) / 10))
        if 'average - CPU - time' in line:
            num = re.findall("\d+",line)[0]
            tmp.append(num)
        if 'the real time is ' in line:
            num = re.findall("\d+",line)[0]
            tmp.append(str(int(num)/10))
            read.append(tmp)
            tmp = []

    print("number totalmax average-CPU real-time")
    data.append(read)
    print(np.asarray(data[-1]))
```

```python
count = 0
res = []
for it in zip(data[0], data[1]):
    tmp1 = float(it[1][3]) - float(it[0][4])
    str_tmp = str(round(tmp1,3))
    tmp = data[1][count]
    tmp.append(str_tmp)
    print(data[1][count])
    count += 1
    print(str_tmp)
    res.append(tmp)
data[1] = res
```
